Aula 3 - Manipulação de Dados e Criação de Gráficos com bibliotecas Python

Lendo o que foi selecionado

```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

Renomeando para tornar a sintaxe mais amigável ao Python

```python
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final', 'Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

Implementando as mesmas operações da planilha com código em Python

```python
df_principal['var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['valor_inicial'] = df_principal['valor_final'] / (1 + df_principal['var_pct'])
df_principal
```

Merge = "Agrupar" as subpastas na planilha df_principal (df = dataframe)

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

Drop, utilizado para excluir uma determinado coluna

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial']) * df_principal['Qtde. Teórica']
df_principal
```

Mudando o formato para float

```python
pd.options.display.float_format = '{:.2f}'.format
```

Deixando a Qtde. Teórica como tipo int

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica': 'qtd_teorica'}).copy()
df_principal
```

Função de linha a linha utilizando lambda

```python
df_principal['resultado'] = df_principal['variacao_rs'].apply(lambda x: "Subiu" if x > 0 else ('Desceu' if x < 0  else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Nome da empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da empresa'])
df_principal
```

```python
df_principal = df_principal.rename(columns={'Ativo':'ativo', 'Data': 'data', 'variacao_rs': 'var_rs', 'Nome': 'nome', 'Segmento': 'segmento', 'Idade': 'idade', 'Cat_Idade': 'cat_idade'}).copy()
df_principal
```

```python
df_principal['cat_idade'] = df_principal['idade'].apply(lambda x: "Mais de 100" if x > 100 else ('Menos de 50' if x < 50  else 'Entre 50 e 100'))
df_principal
```

```python
maior = df_principal['var_rs'].max()
menor = df_principal['var_rs'].min()
media = df_principal['var_rs'].mean()

media_subiu = df_principal[df_principal['resultado'] == "Subiu"]['var_rs'].mean()

media_desceu = df_principal[df_principal['resultado'] == "Desceu"]['var_rs'].mean()

print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Média\tR$ {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['resultado'] == 'Subiu']
df_principal
```

Função de agregação groupby(), somar todo por segmento, "agrupar por". reset_index para continuar como dataframe



```python
df_analise_segmento = df_principal_subiu.groupby('segmento')['var_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('resultado')['var_rs'].sum().reset_index()
df_analise_saldo
```

Criando gráficos, usando a biblioteca Plotly Express

```python
fig = px.bar(df_analise_saldo, x='resultado', y='var_rs', text='var_rs', title='Variação Reais por Resultado')
fig.show()
```

Pesquise com a documentação da biblioteca Plotly ou GPT como mudar a formatação dos números do gráfico de barras

```python
fig = px.bar(df_analise_saldo, x='resultado', y='var_rs', text='var_rs', title='Variação Reais por Resultado')
fig.update_traces(texttemplate='R$ %{text:,.2f}', textposition='outside')
fig.show()
```

Fazer o gráfico de pizza no df_análise_segmentos com a mesma biblioteca Potly.

```python
fig = px.pie(df_analise_segmento, values='var_rs', names='segmento', title='Variação Reais por segmento',)
fig.update_traces(textinfo='value', textposition='inside')
fig.show()
```

Fazer o GroupBy da categoria de idades e gerar o gráfico de barras.

```python
df_analise_cat_idade = df_principal.groupby('cat_idade')['var_rs'].sum().reset_index()
df_analise_cat_idade
```

```python
fig = px.bar(df_analise_cat_idade, x='cat_idade', y='var_rs', text='var_rs', title='Variação Reais por Categoria de Idade')
fig.update_traces(texttemplate='R$ %{text:,.2f}', textposition='outside')
fig.show()
```
